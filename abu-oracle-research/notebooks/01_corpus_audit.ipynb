# Notebook 01 — Corpus Audit

Audits the biographical event corpus used for HF validation.

**Inputs:** `data/corpus/events_public.csv`, `data/corpus/events_summary.csv`  
**Outputs:** `figures/corpus_overview.png`

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('.').resolve()  # adjust if running from a subdirectory

summary = pd.read_csv(REPO / 'data/corpus/events_summary.csv')
events  = pd.read_csv(REPO / 'data/corpus/events_public.csv')

print('Corpus summary:')
print(summary.T)
print()
print('Event types:')
print(events['event_type'].value_counts())
```

```python
# Valence distribution
events['valence'].value_counts().plot.pie(
    colors=['#C9A84C', '#8B2020', '#555555'],
    autopct='%1.1f%%'
)
plt.title('Event Valence Distribution')
plt.show()
```

```python
# HF distribution by valence
events.boxplot(column='hf_weighted', by='valence')
plt.title('HF Weighted by Valence')
plt.suptitle('')
plt.show()
```
